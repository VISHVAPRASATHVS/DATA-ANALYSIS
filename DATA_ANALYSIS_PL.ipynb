```python
t3 = d.groupby("Item Name")["Qty."].sum().sort_values(ascending=False)[0:3] #top 3 items
t3
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(t3.index , t3.values)
plt.xlabel = "item_name"
plt.ylabel = "quantity"
plt.title = "top 3 items"
plt.show()
```

```python
r = d.groupby("Item Name")['Final Total'].sum().sort_values(ascending=False)[0:3] #top 3 items
r
```

```python
plt.bar(r.index , r.values)
plt.xlabel = "item_name"
plt.ylabel = "quantity"
plt.title = "top 3 items"
plt.show()
```
